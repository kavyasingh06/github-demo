```python

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


df = pd.read_csv('songs.csv')
df = df[['title', 'artist', 'genre', 'lyrics']].dropna()


df['combined'] = df.apply(lambda row: f"{row['title']} {row['artist']} {row['genre']} {row['lyrics']}", axis=1)

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['combined'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend(song_title, top_n=5):
    if song_title not in df['title'].values:
        return "❌ Song not found in dataset"
    idx = df[df['title'] == song_title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    song_indices = [i[0] for i in sim_scores[1:top_n+1]]
    return df[['title', 'artist']].iloc[song_indices]


song_name = "Shape of You"  # Change to any song in songs.csv
results = recommend(song_name)
print(f"🎧 Recommended songs because you liked: {song_name}")
print(results)
```

```python
recommend("Believer")
recommend("Perfect")
recommend("Someone Like You")
```
